```python
df["is_prepaid"] = (df["contract_type"]=="Prepaid").astype(int)
df["high_fee"] = (df["monthly_fee"]>10000).astype(int)
df["many_support"] = (df["support_calls"]>3).astype(int)
df["short_tenure"] = (df["tenure_months"]<6).astype(int)
```

```python
df["risk_score"] = (0.4*df["is_prepaid"] + 
                    0.2*df["high_fee"] + 
                    0.25*df["many_support"] + 
                    0.15*df["short_tenure"])
```

```python
print("\nТоп-5 клиентов по риску:")
print(df[["customer_id","contract_type","monthly_fee",
          "support_calls","tenure_months","risk_score","churn"]]
      .sort_values("risk_score", ascending=False).head())
```
