```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/advertising_dataset_regression.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train, y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_predictions)
```

```python
mse
```

```python
model_two = Ridge(alpha=1)
```

```python
model_two.fit(X_train, y_train)
```

```python
y_predictions1 = model_two.predict(X_test)
```

```python
mean_squared_error(y_test, y_predictions1)
```

```python
model_three = Ridge(alpha=0.1)
```

```python
model_three.fit(X_train, y_train)
```

```python
y_predictions2 = model_three.predict(X_test)
```

```python
mean_squared_error(y_test, y_predictions2)
```

```python
model_4 = Ridge(alpha=0.001)
```

```python
model_4.fit(X_train, y_train)
```

```python
y_predictions3 = model_4.predict(X_test)
```

```python
mean_squared_error(y_test, y_predictions3)
```
