```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/advertising_dataset_regression.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_eval = scaler.transform(X_eval)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train, y_train)
```

```python
y_eval_pred = model.predict(X_eval)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_eval, y_eval_pred)
```

```python
mse
```

```python
model1 = Ridge(alpha=1)
```

```python
model1.fit(X_train, y_train)
```

```python
y_eval_predict_1 = model1.predict(X_eval)
```

```python
mse1 = mean_squared_error(y_eval, y_eval_predict_1)
```

```python
mse1
```

```python
model2 = Ridge(alpha=0.1)
```

```python
model2.fit(X_train, y_train)
```

```python
y_eval_predict_2 = model2.predict(X_eval)
```

```python
mse2 = mean_squared_error(y_eval, y_eval_predict_2)
```

```python
mse2
```

```python
y_eval_final = model2.predict(X_test) 
```

```python
mse_final = mean_squared_error(y_test, y_eval_final)
```
