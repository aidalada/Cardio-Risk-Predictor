```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
    np.random.seed(seed)

    sample_ages = np.random.normal(loc=mu, scale=sigma, size=num_samples)
    sample_ages = np.round(sample_ages, decimals=0)

    return sample_ages

sample = create_ages()

sample

sns.kdeplot(data=sample)

sns.displot(data=sample, bins=20)

sns.histplot(data=sample, bins=20, kde=True)

sns.boxplot(x=sample)

ser = pd.Series(data=sample)

ser.describe()

ser[ser<42]

IQR = 55.25000 - 42.00000
lower_limit = 42.00000 - 1.5 * IQR

lower_limit

ser[ser < lower_limit]

higher_limit = 55.25000 + 1.5 * IQR

higher_limit

ser[ser > lower_limit]

np.percentile(sample, [25,75])

q25, q75 = np.percentile(sample, [25,75])

q25

df = pd.read_csv('../../data/Ames_Housing_Data.csv')

df

df.corr(numeric_only=True)['SalePrice'].sort_values()

plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='OverallQual', y='SalePrice')

sns.scatterplot(data=df, x='GrLivArea', y='SalePrice')

df[(df['OverallQual'] > 8) & (df['SalePrice'] < 200000)]

df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 2000000)]

drop_index = df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 200000)].index

df = df.drop(drop_index, axis=0)

sns.scatterplot(data=df, x='GrLivArea', y='SalePrice')

help(pd.to_csv())
```
